````python
img = cv2.imread('night.jpg', cv2.IMREAD_COLOR)

#HSV ìƒ‰ìƒ(Hue), ì±„ë„(Saturation), ëª…ë„(Value) ì±„ë„ë¶„ë¦¬ í›„ value ì±„ë„ë§Œ í‰í‰íˆ í¸ë‹¤.
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
h,s,v = cv2.split(hsv)
v1 = cv2.equalizeHist(v)
hsv1 = cv2.merge((h,s,v1)) #ë³€í˜•ëœ Value ì±„ë„ì„ í•©ì¹¨
img_hsv_eq = cv2.cvtColor(hsv1, cv2.COLOR_HSV2BGR)

plt.axis('off')
plt.imshow(img_hsv_eq)
plt.show()
````

#### ë™ì˜ìƒì—ì„œ ë³´í–‰ì ê²€ì¶œ

- cv2.HOGDescriptor()ì—ì„œ HOGëŠ” Histogram of Gradientë¡œ HOGDescriptor()ëŠ” ê°ì²´ë¥¼ ê²€ì¶œí•˜ê¸° ìœ„í•´ ë§Œë“¤ì–´ì§„ í•¨ìˆ˜ì´ë‹¤.
- setSVMDetector()ì—ì„œ SVMì€ Supprt Vector Machineì˜ ì•½ìë¡œ, ì—¬ë°±ì„ ìµœëŒ€í™”í•˜ì—¬ ì´ˆí‰ë©´ì„ ì°¾ëŠ” ì§€ë„í•™ìŠµ ì•Œê³ ë¦¬ì¦˜ì´ë‹¤.

````python
hog = cv2.HOGDescriptor()
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector())
````

- cv2ëª¨ë“ˆì˜ VideoCapture()ë¡œ ë™ì˜ìƒì„ ë¶ˆëŸ¬ì˜¨ë‹¤.
- ë°˜ë³µë¬¸ì„ í†µí•´ ë™ì˜ìƒì—ì„œ í”„ë ˆì„ì„ ì½ì–´ì˜¨ë‹¤. í”„ë ˆì„ì´ ì œëŒ€ë¡œ ì½í˜”ë‹¤ë©´ ì¦‰ ret==Trueì´ë¼ë©´, ê·¸ í”„ë ˆì„ì—ì„œ ë³´í–‰ìë¥¼ ê²€ì¶œí•œë‹¤.

````python
from IPython.display import clear_output

cap = cv2.VideoCapture('walking.avi')

while True:
    ret, frame = cap.read()
    if not ret:
        break
    detected, _ = hog.detectMultiScale(frame) # ë§¤ í”„ë ˆì„ë§ˆë‹¤ ë³´í–‰ì ê²€ì¶œ

    # ê²€ì¶œ ê²°ê³¼ í™”ë©´ í‘œì‹œ
    # ì‚¬ê°í˜• ì‹œì‘ì¢Œí‘œ x,y, ì¢…ë£Œì¢Œí‘œ w,h, ì‚¬ê°í˜•ìƒ‰ê¹”(127,0,0) , ì‚¬ê°í˜• ë‘ê»˜ 2
    for (x, y, w, h) in detected:
        cv2.rectangle(frame, (x, y, w, h), (127, 0, 0), 2)

    clear_output(wait=True)
    cv2_imshow(frame)
    # if cv2.waitKey(10) == 27:
    #     break
````

## ìì—°ì–´ ì²˜ë¦¬ ìƒ˜í”Œ ì½”ë“œ

````python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.tag import pos_tag
from nltk.chunk import ne_chunk

# NLTK ë°ì´í„° ë‹¤ìš´ë¡œë“œ (ìµœì´ˆ ì‹¤í–‰ ì‹œ í•„ìš”)
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')

# ì…ë ¥ í…ìŠ¤íŠ¸
text = "Apple is looking at buying U.K. startup for $1 billion. Microsoft was founded by Bill Gates and Paul Allen."

# ë¬¸ì¥ í† í°í™”
sentences = sent_tokenize(text)
print("ë¬¸ì¥ í† í°í™”:")
print(sentences)

# ë‹¨ì–´ í† í°í™” ë° í’ˆì‚¬ íƒœê¹…
for sentence in sentences:
    words = word_tokenize(sentence)
    tagged_words = pos_tag(words)
    print("\në‹¨ì–´ í† í°í™” ë° í’ˆì‚¬ íƒœê¹…:")
    print(tagged_words)

    # ëª…ëª…ëœ ê°œì²´ ì¸ì‹
    named_entities = ne_chunk(tagged_words)
    print("\nëª…ëª…ëœ ê°œì²´ ì¸ì‹:")
    print(named_entities)
````

````python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

# ëª¨ë¸ê³¼ í† í¬ë‚˜ì´ì €ë¥¼ ë¡œë“œí•©ë‹ˆë‹¤.
model_name = 'gpt2'  # 'gpt2-medium', 'gpt2-large', 'gpt2-xl' ë“±ìœ¼ë¡œ ë³€ê²½ ê°€ëŠ¥
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name)

# ì…ë ¥ í…ìŠ¤íŠ¸ ì„¤ì •
input_text = "Once upon a time"

# ì…ë ¥ í…ìŠ¤íŠ¸ë¥¼ í† í°í™”í•©ë‹ˆë‹¤.
input_ids = tokenizer.encode(input_text, return_tensors='pt')

# í…ìŠ¤íŠ¸ë¥¼ ìƒì„±í•©ë‹ˆë‹¤.
output = model.generate(input_ids, max_length=100, num_return_sequences=1)

# ìƒì„±ëœ í…ìŠ¤íŠ¸ë¥¼ ë””ì½”ë”©í•©ë‹ˆë‹¤.
generated_text = tokenizer.decode(output[0], skip_special_tokens=True)

print("ìƒì„±ëœ í…ìŠ¤íŠ¸:")
print(generated_text)
````
