## return value(반환값)
- return의 역할: 
  - result를 호출자에게 반환
  - 함수 탈출
- return 이후에는 어떠한 코드가 있어도 전혀 실행되지 않음

````python
# mySum()함수 정의해보기

def mySum():
  total=0
  for num in range(1,101):
      total = total + num
  return total #함수안에서 나온 결과를 함수 밖에서 사용하고자 할때에는 반드시 return이 있어야 함

print(mySum()) # return이 없을때와 있을 때 출력값 차이
````

## 변수의 유효 범위( Variable scope)




Global variable vs. local variable

````python
def func1():
    v1=50 #local variable

def func2():
    v2=70 #local variable

func1()  
print(v1)
````

````python
def func1():
    v1=50
    print(v2)

def func2():
    v2=70
    func1() # func1() 호출

func2() # func2() 호출
````

````python
v2 = 3 #global variable
def func1():
    global v2
    v1=50
    v2 = v2 + 1
    print(v2)

def func2():
    v2=70
    print(v2)
    func1()

func2()
print(v2)
````
