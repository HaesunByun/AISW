## Numpy ì‘ìš©: ë°ì´í„° ë¶„ì„

ë°°ì—´ì˜ ì‘ìš©
- ë°ì´í„° ë¶„ì„ ì˜ˆ: ë‚ ì”¨ ë°ì´í„°ë¥¼ ë¶„ì„. íŠ¹ì • ê¸°ê°„ ë™ì•ˆì˜ ì˜¨ë„ ë°ì´í„°ë¥¼ ë¶„ì„í•´ í‰ê· , ìµœëŒ€, ìµœì†Œ ê°’ì„ ê³„ì‚°

````python
import numpy as np

# ì˜ˆì œ: ì¼ì£¼ì¼ê°„ì˜ ì˜¨ë„ ë°ì´í„° (ì„­ì”¨)
temperatures = np.array([22.1, 23.4, 21.8, 22.9, 23.0, 24.2, 23.5])

# í‰ê·  ì˜¨ë„
mean_temp = np.mean(temperatures)

# ìµœëŒ€ ì˜¨ë„
max_temp = np.max(temperatures)

# ìµœì†Œ ì˜¨ë„
min_temp = np.min(temperatures)

print(f"í‰ê·  ì˜¨ë„: {mean_temp:.2f}Â°C")
print(f"ìµœëŒ€ ì˜¨ë„: {max_temp:.2f}Â°C")
print(f"ìµœì†Œ ì˜¨ë„: {min_temp:.2f}Â°C")
````
