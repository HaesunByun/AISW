## Deep Copy vs. Shallow Copy
ë°°ì—´ì˜ ë°ì´í„°ë¥¼ ë³µì‚¬í•´ì„œ ë³µì‚¬ëœ ë°ì´í„°ë¥¼ ê°€ì§€ê³  ì‘ì—…ì„ í•´ì•¼ í•  ë•Œê°€ ìˆë‹¤.
ë°°ì—´ì„ ë³µì‚¬í•˜ëŠ” ë°©ë²•ì—ëŠ” ë‘ ê°€ì§€ê°€ ìˆë‹¤.

- Shallow Copy : ì›ì¡°ë³€ìˆ˜ì™€ ë³µì‚¬ë³€ìˆ˜ê°€ ê°™ì€ ë©”ëª¨ë¦¬ë¥¼ ì°¸ì¡°
- Deep Copy : ë³µì‚¬ë³€ìˆ˜ëŠ” ìƒˆë¡œì´ ë©”ëª¨ë¦¬ë¥¼ í• ë‹¹í•˜ì—¬ ë°°ì—´ì„ ìƒì„±

- `cp_ar = ar`ì™€ ê°™ì€ ë³µì‚¬ë¥¼ shallow copyë¼ê³  í•œë‹¤.
- shallow ë³µì‚¬ì˜ ë¬¸ì œëŠ” ë³µì‚¬ ë°°ì—´ì˜ ë°ì´í„°ë¥¼ ì‘ì—…ì„ í•˜ë©´ ì›ì¡° ë°°ì—´ì˜ ë°ì´í„°ë„ ë˜‘ê°™ì´ ì‘ì—…ëœë‹¤ëŠ” ê²ƒì´ë‹¤.
- ê·¸ ì´ìœ ëŠ” ë³µì‚¬ë³€ìˆ˜ cp_arê³¼ ì›ì¡°ë³€ìˆ˜ arì´ ê°™ì€ ë©”ëª¨ë¦¬ ì£¼ì†Œë¥¼ ì°¸ì¡°í•˜ê¸° ë•Œë¬¸ì´ë‹¤.

````python
import numpy as np
ar = np.arange(1,10).reshape(3,3)
print(ar)

cp_ar = ar  #shallow copy
print(cp_ar)

cp_ar[0][2]=1000
print(cp_ar,'\n\n', ar) # cp_arì˜ ê°’ì„ ë°”ê¿¨ëŠ”ë° arë„ ë°”ë€Œì–´ ìˆìŒ
````


- ì´ë¥¼ ë°©ì§€í•˜ê¸° ìœ„í•´ì„œëŠ” cp_ar = ar.copy()ì™€ ê°™ì´ Deep Copyë¥¼ í•´ì•¼ í•œë‹¤.
- Deep copyëŠ” ë³µì‚¬ë³€ìˆ˜ì˜ ê°’ì„ ì‘ì—…(ìˆ˜ì •, ì‚­ì œ, ì¶”ê°€ ë“±)í•´ë„ ì›ë³¸ë³€ìˆ˜ì˜ ê°’ì´ ì‘ì—…ë˜ì§€ ì•ŠëŠ”ë‹¤.

````python
ar = np.arange(1,10).reshape(3,3)
print(ar)

cp_ar = ar.copy() #Deep copy
print(cp_ar)

cp_ar[0][2]=1000
print(cp_ar,'\n\n', ar) # ë³µì‚¬ëœ ê³³ì—ì„œ ê°’ì„ ìˆ˜ì •í•´ë„ ì›ë³¸ì— ì˜í–¥ì—†ìŒ
````
